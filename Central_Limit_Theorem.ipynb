```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
# Подключаем нужные библиотеки
```

```python
# Создаем выборку объемом 1000 распределения Гилбрата
# Строим по ней гистограмму, а поверх нее функцию плотности распределения
sample = sts.gilbrat.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
x = np.linspace(sts.gilbrat.ppf(0.001), sts.gilbrat.ppf(0.999), 100)

ax.plot(x, sts.gilbrat.pdf(x),'r-', lw=5, alpha=0.6, label='gilbrat pdf')
plt.hist(sample, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

```python
# Строим гистограмму выборочного среднего для выборки объема 5
# Поверх нее строим функцию распределения плотности вероятности нормального распределения
fig, ax = plt.subplots(1, 1)
n = 5
# Матожидание распределения Гилбрата = (e) ^ 0.5
mean = math.sqrt(math.e)
# Среднее отклонение распределения Гилбрата = e (e - 1)
var = math.sqrt(math.e * (math.e - 1) / n)

x = np.linspace(sts.norm.ppf(0.001), sts.norm.ppf(0.99999), 100)
ax.plot(x, sts.norm.pdf(x, mean, var), 'r-', lw=5, alpha=0.6, label='norm pdf')

sample = []
for i in range(1000):
    selection = sts.gilbrat.rvs(size=n)
    avg = sum(selection)/n
    sample.append(avg)
plt.hist(sample, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```
